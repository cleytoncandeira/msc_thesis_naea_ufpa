## Step 2 -- Extract audit companies from .PDFs files

```python
import PyPDF2
import camelot
import re

path = "/home/cleyton/ProjetosGit/msc_thesis_naea_ufpa/audit_reports/"
file_name = "audit_report_rtrs"
```

```python
#how we found the correct page that we'll extract tables?

auditing_companies_tags = {"Bureau Veritas": 'BUREAU', 
          "Control Union": 'CUC', 
          "FoodChain Id Certificadora Ltda": 'FCID',  
          "Gênesis Certificações": '-GC', 
          "SGS Group": 'SGS',
          "Schutter Argentina": 'SCHUT',
          "Cert Id Certificadora": 'CERT ID'
                          }  

def txt_enclosure(txt_file):
    def find_all_re(search):
        return re.findall(search, txt_file)
    return find_all_re

auditing_companies = []
teste = []
for reports_range in range(1, 326):
    
    with open(path + file_name + str(reports_range) + ".pdf", 'rb') as pdf_files:

        #identify .pdf file and classify your report model
        join_pages = ""

        for i in range(1, 4):
            pdf_reader = PyPDF2.PdfReader(pdf_files)
            page_target = pdf_reader.pages[i]
            txt_ext = page_target.extract_text()
            join_pages += txt_ext 
            
        #find all and classification
        general_txt_enclosure = txt_enclosure(join_pages)
        
        
        for auditing_company, tags in auditing_companies_tags.items():
            if general_txt_enclosure(tags):
                auditing_companies.append(auditing_company)
            
```
