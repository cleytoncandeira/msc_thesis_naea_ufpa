```python
import pandas as pd
from pathlib import Path
from extract_audit_companies import auditing_companies_tags

#root 
ROOT_FOLDER = Path('__file__').parent

#import csv
data_base = pd.read_csv(ROOT_FOLDER / 'certified_producers_rtrs.csv')

#adjust index file
data_base.set_index(pd.Index(data_base.iloc[:, 0] + 1, name = "idx_file"), inplace = True)
data_base = data_base.iloc[:, 1:]

#get company names
auditing_companies = [auditing_companies for auditing_companies in auditing_companies_tags.keys()]

#get indexs companies
db_list = [data_base[data_base["Companies"] == aud_comp] for aud_comp in auditing_companies]
    
```

```python
import PyPDF2
import camelot
import re

def read_pg_pdf(pdf_file, start_page, end_page):
    
    join_pages = ""
    
    for page in range(start_page, end_page + 1):
        
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        page_target = pdf_reader.pages[page]
        txt_ext = page_target.extract_text()
        join_pages += txt_ext 
        
    return join_pages
```

```python
file_names = "audit_report_rtrs" + str(1) + ".pdf"
            
open_way_reports = ROOT_FOLDER / 'audit_reports' / file_names


with open(open_way_reports, 'rb') as report:
    txt = read_pg_pdf(report, 1, 9)
    print(txt)
    
    
```
