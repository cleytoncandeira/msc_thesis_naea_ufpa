# Natural Language Processing for RTRS DataBase

## Cleaning Data

```python
import pandas as pd
import numpy as np
```

```python
rtrs_database = pd.read_csv('./csv_files/brazil_rtrs.csv')
n_rows, n_cols = rtrs_database.shape 
col_names = rtrs_database.columns.to_list()
col_names = col_names[-28:]
```

```python
#%pip install nltk

import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from unicodedata import normalize
```

```python
def tratamento_texto(txt):
    return ' '.join((remover_especiais_str(txt).split()))
                    
def remover_especiais_str(txt):
    return re.sub(r'[^a-zA-Z0-9 ]', ' ', normalize('NFKD', txt).encode('ASCII', 'ignore').decode('ASCII'))
```

```python
colslists = list()

for col in col_names:
	clean_col = list()
	for i in range(n_rows):
		corpus = tratamento_texto(rtrs_database[col][i])
		corpus = corpus.lower()
		corpus = corpus.split()
		ps = PorterStemmer()
		corpus = [ps.stem(word) for word in corpus if not word in set(stopwords.words('portuguese'))]
		corpus = ' '.join(corpus)
		clean_col.append(corpus)
	colslists.append(clean_col)	
```
